## **Which are the most populated asian countries in all three datasets by the year 2022?**

## **What is the distribution of gender in the most populated asian countries?**

### Reframing the dataset to work upon easily:
- ### Creating only 3 columns : [Country, Year, Population]
- ### Plotting boxplots for male and female from 1960-2022 to understand the distribution of gender over the years and across the countries

```python
## segregating the years from the df_male_asian
years = df_male_asian.select_dtypes([int, float]).columns

## list of top 5 asian countries
country_names = ["India", "China", "Indonesia", "Pakistan", "Bangladesh"]

## creating a list of population for each consecutive countries
male_country_population = [
    df_male_asian_copy.iloc[0, 1:].tolist(),
    df_male_asian_copy.iloc[1, 1:].tolist(),
    df_male_asian_copy.iloc[2, 1:].tolist(),
    df_male_asian_copy.iloc[3, 1:].tolist(),
    df_male_asian_copy.iloc[4, 1:].tolist()
]

maledata = []

for country, male_population in zip(country_names, male_country_population):
    maledata.extend([{"Country" : country, "Year" : year, "Population" : malepop} for year, malepop in zip(years, male_population)])

male_data = pd.DataFrame(maledata)


female_country_population = [
    df_female_asian_copy.iloc[0, 1:].tolist(),
    df_female_asian_copy.iloc[1, 1:].tolist(),
    df_female_asian_copy.iloc[2, 1:].tolist(),
    df_female_asian_copy.iloc[3, 1:].tolist(),
    df_female_asian_copy.iloc[4, 1:].tolist()
]

femaledata = []

for country, femalepopulation in zip(country_names, female_country_population):
    femaledata.extend([{"Country" : country, "Year" : year, "Population" : femalepop} for year, femalepop in zip(years, femalepopulation)])

female_data = pd.DataFrame(femaledata)

print(male_data.head())                     #For each country it will store the population data from 1960-2022 and repeat till 5th countries
print('\n',female_data.head())
```

### Plotting Boxplots:

```python
## setting up the y_ticks and y_ticklabel to match the real data

min_population = 24795178.0
max_population = 731180498.0

num_ticks = 15

y_ticks = np.linspace(min_population, max_population, num_ticks).astype(int)    ## using linspace to divide the min and max population equally

y_tick_labels = [f"{val / 1000000:.2f}M" for val in y_ticks]                    ## converting the y_ticks value into Million to be used as y_ticklabel

# --------------------------------------------------------------------------------------------------------------------------------------- #

fig, axes = plt.subplots(1,2, figsize = (15,6))

sns.set_style("whitegrid")
sns.boxplot(x = male_data["Country"], y = male_data["Population"], hue = male_data["Country"], ax = axes[0])
axes[0].set_title("Boxplot for Male Asian Population")
axes[0].set_yticks(y_ticks)
axes[0].set_yticklabels(y_tick_labels)


sns.boxplot(x = female_data["Country"], y = female_data["Population"], hue = female_data["Country"], ax = axes[1])
axes[1].set_title("Boxplot for Female Asian Population")
axes[1].set_yticks(y_ticks)
axes[1].set_yticklabels(y_tick_labels)


plt.tight_layout()
plt.show()
```

# **Plotting Histogram for each country based on Male and Female Population**:

```python
# Add a Gender column to both datasets
male_data["Gender"] = "M"
female_data["Gender"] = "F"

# Merge the datasets
merged_data = pd.concat([male_data, female_data], ignore_index=True)

# Convert the "Year" column to integers
merged_data["Year"] = merged_data["Year"].astype(int)

# Plot histograms
# The loop iterates through each country, creating a histogram for the population data
# in each subplot, with specific formatting for the title, x-axis ticks, and x-axis tick labels.

sns.set(style="whitegrid")
fig, axes = plt.subplots(3, 2, figsize=(15, 10))

for (country, ax) in zip(merged_data["Country"].unique(), axes.flatten()):                                          # This line iterates over pairs of unique countries and corresponding subplot axes. merged_data["Country"].unique() gets unique country names, and axes.flatten() flattens the 2D array of subplot axes into a 1D array
    data = merged_data[merged_data["Country"] == country]                                                           # This line extracts the data for the current country from the merged dataset.
    sns.histplot(data=data, x="Population", bins=8, kde=True, ax=ax, color="blue", hue="Gender", palette="rocket")  # This line creates a histogram plot using Seaborn for the population data of the current country. The ax=ax parameter specifies the subplot where the plot should be placed

    ax.set_title(f"Histogram for {country} Population Gender-wise")
    ax.set_xticks(np.linspace(24795178, 731180498, 9).astype(int))                                                  # This line sets the x-axis ticks to specific values (in this case, population values). It creates 8 ticks spaced evenly between 24,795,178 and 731,180,498.
    ax.set_xticklabels([f"{val/1000000:.2f}M" for val in np.linspace(24795178, 731180498, 9).astype(int)])          # This line sets the labels for the x-axis ticks, converting the tick values to millions with two decimal places

plt.tight_layout()
plt.show()
```

### Observations From Above Charts:

- **Population Distribution Across Countries:**
  - Countries like *India* and *China* exhibit population spreads that surpass others, such as *Indonesia*, *Pakistan*, and *Bangladesh*.

- **India:**
  1. The *male population* shows a slight peak between 200-300 million, with an overall evenly distributed trend.
  2. The *female population* has more pronounced peaks between 500-650 million, displaying a positively skewed distribution.

- **China:**
  1. The *male population* is highly right-skewed, with the majority falling between 500-700 million.
  2. The *female population* has slight peaks around 200-280 million, showing an overall evenly distributed pattern.

- **Indonesia, Pakistan, Bangladesh:**
  - The population in these countries exhibits slight crests and troughs, with an overall distribution between 24-112 million.

*Note: These observations are based on the histograms depicting population distribution by gender for each country.*


# Plotting Bar Graph to tally the gender-wise distribution among countries in the year 1960 and 2022:

```python
# Defining a function plot_population_distribution to encapsulate the common plotting logic for both years
# create bar plots for the gender-wise population distribution in 1960 and 2022.
def plot_population_distribution(data, year, ax):
    sns.set(style="darkgrid")
    graph = sns.barplot(x=data["Country"], y=data["Population"], hue=data["Gender"], ax=ax, palette="rocket")
    population_ticks = np.linspace(data["Population"].min(), data["Population"].max(), 13)
    graph.set_yticks(population_ticks)
    graph.set_yticklabels([f"{pop/1000000:.2f}M" for pop in population_ticks])
    graph.set_title(f"Gender-Wise Population distribution in {year}")

# Filter data for the year 1960
data_1960 = merged_data[merged_data["Year"] == 1960]

# Filter data for the year 2022
data_2022 = merged_data[merged_data["Year"] == 2022]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plot population distribution for 1960
plot_population_distribution(data_1960, 1960, axes[0])

# Plot population distribution for 2022
plot_population_distribution(data_2022, 2022, axes[1])

plt.tight_layout()
plt.show()
```

3. **China:**
   - China's female population was much lower than its male population:
     - Female Population: Approx. 207 million
     - Male Population: Approx. 340 million

4. **Indonesia, Pakistan, Bangladesh:**
   - Male-female ratios in these countries were somewhat equal, with no major disparities.

#### In 2022:

1. **Population Increase:**
   - The population of all five Asian countries in 2022 increased and remained under 740 million.

2. **India:**

3. **China:**

```python
countries = merged_data["Country"].unique()

fig, axes = plt.subplots(3, 2, figsize=(15, 9))
sns.set_style("dark")

# The loop processes each country, creating a line plot for the population over
# the years in a subplot, with customized ticks, labels, and titles
for i, country in enumerate(countries):                                                                                                   # iterates over each unique country in the countries array, and i is the index of the loop.
    data = merged_data[merged_data["Country"] == country]                                                                                 # "Country" column matches the current country in the loop & creates a subset of data specific to each country
    row, col = divmod(i, 2)                                                                                                               # divmod - used to calculate (quot.,rem.) of the division of i by 2 which is used to convert a linear index (i) into row and column indices (row and col) for a 2D grid, it's used to determine the subplot position

    sns.lineplot(x="Year", y="Population", data=data, ax=axes[row, col], hue="Gender")
    axes[row, col].set_title(country)

    # Set x-axis ticks and labels
    axes[row, col].set_xticks(np.linspace(data["Year"].min(), data["Year"].max(), 15))
    axes[row, col].set_xticklabels([str(int(year)) for year in np.linspace(data["Year"].min(), data["Year"].max(), 15)])

    # Set y-axis ticks and labels
    axes[row, col].set_yticks(np.linspace(data["Population"].min(), data["Population"].max(), 6))
    axes[row, col].set_yticklabels([f"{pop/1000000:.2f}M" for pop in np.linspace(data["Population"].min(), data["Population"].max(), 6)])

    # Adjust x-axis limits based on the data range for the specific country
    axes[row, col].set_xlim(data["Year"].min(), data["Year"].max())

plt.grid(False)
plt.tight_layout()
plt.show()
```

### Observation From Above Charts:

- All the above countries have seen an increasing trend in their population numbers from 1960 to 2022.

#### India:

2. After 2013, the trend shifted, and the male population surpassed the female population, continuing to the present.

#### China:
1. In China, the male population has consistently been higher than the female population since the year 1960.

#### Indonesia, Pakistan, Bangladesh:
1. Both male and female populations in Indonesia, Pakistan, and Bangladesh have shown an overall increase from 1960 to 2022.
2. The growth patterns for male and female populations in these countries are similar over the observed period.
